```python
import codecs
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

import sys
sys.path.append('../../')
from exact_barycenter import *
from softmax import *
from scipy.optimize import minimize
from misc import *
from knn import *
```

```python
def read_data(filename):
    dataFile = codecs.open(filename)
    text = dataFile.readlines()
    
    data = []
    labels = []

    for line in text:
        new = [float(x) for x in line.split()[1:]]
        data.append(new)
        labels.append(float(line.split()[0]))

    data = np.array(data)
    labels = np.array(labels)
    ys = np.array(labels) -1

    support, psds = signal.periodogram(data)
    
    normalize = np.sum(psds, axis=1)
    psds = (psds.transpose() / normalize).transpose()
    
    return data, psds, ys, support

data_test, X_te, y_te, support = read_data('Data/Fungi/Fungi_TEST.txt')
data_train, X_tr, y_tr, _ = read_data('Data/Fungi/Fungi_TRAIN.txt')

X = np.vstack((X_tr, X_te))
y = np.hstack((y_tr, y_te))
X += 1e-30
```

```python
print(X.shape)
```

```python
n_class = len(np.unique(y))
x0=np.ones((n_class+1)*n_class)*1e-3
options_q = {'q_support': 1000}
seeds = np.arange(10)
methods = ['quantile', 'euclidean', 'KL']
scores = np.zeros((len(seeds), len(methods)))

for s in seeds:
    for j, m in enumerate(methods):
        X_tr, X_te, y_tr, y_te = random_split(X, y, 0.8, seed=s)
        y_tr = y_tr.reshape(-1,)
        
        model_sink = WSoftmax(X_tr, y_tr, support, distance=m, quadratic=False, options=options_q)
        res = minimize(fun=model_sink.logp, x0=x0, jac=model_sink.dlogp, method='L-BFGS-B')
        theta = res['x']
        prob, pred = model_sink.pred(X_te, theta)
        scores[s, j] = np.sum(pred.reshape(y_te.shape) == y_te) / len(y_te)
```

```python
print(np.mean(scores, axis=0))
print(np.std(scores, axis=0) / np.sqrt(len(seeds)))
```
